```python
#Create a Heatmap : Host departments VS Combined schools

# Create a contingency table (cross-tabulation)
contingency_table = pd.crosstab(df_summarized['Host departments'], df_summarized['Combined schools'])

# Plot heatmap
plt.figure(figsize=(15, 12))
sns.heatmap(contingency_table, annot=True,cmap='PuBu')
plt.title('Heatmap: Host departments vs Combined schools')
plt.xlabel('Combined schools')
plt.ylabel('Host departments')
plt.show()
```

```python
# Create a Heatmap : Host departments VS Combined schools by years

def create_heatmap(year):
    # Assuming you have a DataFrame for each year, modify this line accordingly
    df_year = df_summarized[df_summarized['Year'] == year]
    
    # Create a contingency table (cross-tabulation)
    contingency_table = pd.crosstab(df_year['Host departments'], df_year['Combined schools'])

    # Plot heatmap
    plt.figure(figsize=(15, 12))
    sns.heatmap(contingency_table, annot=True,cmap='PuRd')
    plt.title(f'Heatmap: Host departments vs Combined schools ({year})')
    plt.xlabel('Combined schools')
    plt.ylabel('Host departments')
    plt.show()

# Input year
year_input = input("Enter the year: ")

# Call the function with the input year
create_heatmap(year_input)
```
